# Swarmalator Dynamics

Agents with both spatial position and oscillator phase.
Phase modulates spatial attraction; proximity modulates phase coupling.

O'Keeffe, Hong, Strogatz, Nature Communications 2017.

This notebook explores how different J (phase→space) and K (space→phase)
parameters produce different collective states.

```python
import numpy as np

from scpn_phase_orchestrator.upde.swarmalator import SwarmalatorEngine
```

```python
N = 30
rng = np.random.default_rng(42)
pos0 = rng.uniform(-2, 2, (N, 2))
phases0 = rng.uniform(0, 2 * np.pi, N)
omegas = np.zeros(N)  # identical frequencies
```

## State 1: Static Sync (J > 0, K > 0)

Phase-similar agents attract spatially. Nearby agents synchronize.
Result: clustered, phase-locked.

```python
engine = SwarmalatorEngine(N, 2, dt=0.01, A=1.0, B=1.0, J=1.0, K=1.0)
fp, fph, _, _ = engine.run(pos0.copy(), phases0.copy(), omegas, 2000)

print(f"Phase coherence R: {engine.phase_coherence(fph):.3f}")
print(f"Spatial compactness: {engine.spatial_coherence(fp):.3f}")
print(f"Phase-space correlation: {engine.phase_spatial_correlation(fp, fph):.3f}")
```

## State 2: Static Async (J < 0, K > 0)

Phase-different agents attract. Nearby agents synchronize.
Result: spatial order by phase difference.

```python
engine2 = SwarmalatorEngine(N, 2, dt=0.01, A=1.0, B=1.0, J=-0.5, K=1.0)
fp2, fph2, _, _ = engine2.run(pos0.copy(), phases0.copy(), omegas, 2000)

print(f"Phase coherence R: {engine2.phase_coherence(fph2):.3f}")
print(f"Spatial compactness: {engine2.spatial_coherence(fp2):.3f}")
print(f"Phase-space correlation: {engine2.phase_spatial_correlation(fp2, fph2):.3f}")
```

## State 3: Decoupled (J = 0)

Phase doesn't affect spatial dynamics. Standard swarm + standard Kuramoto.

```python
engine3 = SwarmalatorEngine(N, 2, dt=0.01, A=1.0, B=1.0, J=0.0, K=1.0)
fp3, fph3, _, _ = engine3.run(pos0.copy(), phases0.copy(), omegas, 2000)

print(f"Phase coherence R: {engine3.phase_coherence(fph3):.3f}")
print(f"Spatial compactness: {engine3.spatial_coherence(fp3):.3f}")
print(f"Phase-space correlation: {engine3.phase_spatial_correlation(fp3, fph3):.3f}")
```

## 3D Swarmalators

```python
pos3d = rng.uniform(-2, 2, (N, 3))
engine3d = SwarmalatorEngine(N, 3, dt=0.01, A=1.0, B=1.0, J=0.5, K=1.0)
fp3d, fph3d, _, _ = engine3d.run(pos3d, phases0.copy(), omegas, 1000)

print(f"3D Phase coherence: {engine3d.phase_coherence(fph3d):.3f}")
print(f"3D Spatial compactness: {engine3d.spatial_coherence(fp3d):.3f}")
```
