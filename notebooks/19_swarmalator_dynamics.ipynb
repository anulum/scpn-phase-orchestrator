# Swarmalator Dynamics

Agents with both spatial position and oscillator phase.
Phase modulates spatial attraction; proximity modulates phase coupling.

O'Keeffe, Hong, Strogatz, Nature Communications 2017.

This notebook explores how different J (phase→space) and K (space→phase)
parameters produce different collective states.

```python
import numpy as np

from scpn_phase_orchestrator.upde.swarmalator import SwarmalatorEngine
```

```python
N = 30
rng = np.random.default_rng(42)
pos0 = rng.uniform(-2, 2, (N, 2))
phases0 = rng.uniform(0, 2 * np.pi, N)
omegas = np.zeros(N)  # identical frequencies
```

## State 1: Static Sync (J > 0, K > 0)

Phase-similar agents attract spatially. Nearby agents synchronize.
Result: clustered, phase-locked.

```python
engine = SwarmalatorEngine(N, dim=2, dt=0.01)
fp, fph, _, _ = engine.run(pos0.copy(), phases0.copy(), omegas, a=1.0, b=1.0, j=1.0, k=1.0, n_steps=2000)

R = engine.order_parameter(fph)
mean_dist = np.mean(np.linalg.norm(fp - fp.mean(axis=0), axis=1))
print(f"Phase coherence R: {R:.3f}")
print(f"Mean distance from centroid: {mean_dist:.3f}")
```

## State 2: Static Async (J < 0, K > 0)

Phase-different agents attract. Nearby agents synchronize.
Result: spatial order by phase difference.

```python
engine2 = SwarmalatorEngine(N, dim=2, dt=0.01)
fp2, fph2, _, _ = engine2.run(pos0.copy(), phases0.copy(), omegas, a=1.0, b=1.0, j=-0.5, k=1.0, n_steps=2000)

R2 = engine2.order_parameter(fph2)
mean_dist2 = np.mean(np.linalg.norm(fp2 - fp2.mean(axis=0), axis=1))
print(f"Phase coherence R: {R2:.3f}")
print(f"Mean distance from centroid: {mean_dist2:.3f}")
```

## State 3: Decoupled (J = 0)

Phase doesn't affect spatial dynamics. Standard swarm + standard Kuramoto.

```python
engine3 = SwarmalatorEngine(N, dim=2, dt=0.01)
fp3, fph3, _, _ = engine3.run(pos0.copy(), phases0.copy(), omegas, a=1.0, b=1.0, j=0.0, k=1.0, n_steps=2000)

R3 = engine3.order_parameter(fph3)
mean_dist3 = np.mean(np.linalg.norm(fp3 - fp3.mean(axis=0), axis=1))
print(f"Phase coherence R: {R3:.3f}")
print(f"Mean distance from centroid: {mean_dist3:.3f}")
```

## 3D Swarmalators

```python
pos3d = rng.uniform(-2, 2, (N, 3))
engine3d = SwarmalatorEngine(N, dim=3, dt=0.01)
fp3d, fph3d, _, _ = engine3d.run(pos3d, phases0.copy(), omegas, a=1.0, b=1.0, j=0.5, k=1.0, n_steps=1000)

R3d = engine3d.order_parameter(fph3d)
mean_dist3d = np.mean(np.linalg.norm(fp3d - fp3d.mean(axis=0), axis=1))
print(f"3D Phase coherence: {R3d:.3f}")
print(f"3D Mean distance from centroid: {mean_dist3d:.3f}")
```
